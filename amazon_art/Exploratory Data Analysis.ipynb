# First Model, Artist vs Price

```python
print amazon_art.describe()
amazon_art.loc[amazon_art.price_sqin < 2, 'price_class'] = 0
amazon_art.loc[amazon_art.price_sqin >= 2, 'price_class'] = 1
```

```python
from sklearn.cross_validation import train_test_split

X = artist_df.loc[:,:]
y = amazon_art.loc[:, 'price_class']

X_train, X_test, y_train, y_test = train_test_split( X, y, random_state=1)

print X.shape, y.shape
print X_train.shape, y_train.shape
print X_test.shape, y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn import metrics


logreg = LogisticRegression()  
logreg.fit(X_train, y_train)
preds = logreg.predict(X_test)
```

```python
logreg.score(X_train, y_train)
print metrics.accuracy_score(y_test, preds)
metrics.confusion_matrix(y_test, preds)
```

```python
#Null Accuracy Rate
y_test.mean()
1 - y_test.mean()
```

```python
# Let's graph the accuracy of a few values of K
scores = []
for i in range(1, 30):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X, y)
    scores.append(knn.score(X, y))
    
scores
```

```python
plt.plot(range(1,30), scores)
plt.ylabel("Accuracy")
plt.xlabel("K")
```

# Image extraction from urls

```python
urls = []

for url in amazon_art.image:
    urls.append(url)
    
urls[0]
```

```python
urls = []

for url in amazon_art.image:
    url = url.replace('"','').replace("'",'').replace('uhttp://', 'http://').strip()
    urls.append(url)
    
def url_to_image(url):
    #download the image
    resp = urllib.urlopen(url)
    #convert to NumPy array
    image = np.asarray(bytearray(resp.read()), dtype ='uint8')
    #read it into OpenCV
    image = cv2.imdecode(image, cv2.IMREAD_COLOR)
    
    return image

img_array = []
for url in urls:
    image = url_to_image(url)
    cv_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    img_array.append(cv_rgb)
```
